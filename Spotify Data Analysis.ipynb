**_Interpretation:_**
* The x-axis shows popularity scores (0–100).
* The y-axis shows how many songs fall into each range.
* Common result: Most tracks will have low-to-moderate popularity (0–40), with fewer tracks at the high end.
* This shows right-skewed distribution, meaning only a few songs are very popular — consistent with real-world listening behavior.

### Plot Of Loundness Vs Energy 

```python
sample_data = sp_tracks[["energy", "loudness"]].sample(frac=0.05) # Takes a 5% random sample of energy and loudness columns from sp_tracks.
```

**_Interpretation:_**
* Sampling speeds up plotting and reduces overplotting in scatter/regression visuals. The sample preserves overall trends while making visuals manageable.

```python
# Scatterplot of Energy and Loudness
plt.figure(figsize= (10,6))
plt.scatter(x= sample_data["energy"], y = sample_data["loudness"], edgecolor = "k")
plt.title("Scatterplot of Energy and Loudness")
plt.xlabel("Energy")
plt.ylabel("Loudness")
plt.grid()
plt.show
```

**_Explanation_**

* Takes a 5% random sample to make the plot manageable (sample(frac=0.05)).

* Scatter plot compares energy (0–1) vs loudness (in decibels).

* alpha=0.5 adds transparency to prevent overlap.

**_Interpretation_**

* You’ll likely see a positive correlation: higher energy tracks are louder.

* This reflects real-world physics and production: energetic songs are mixed louder.

* Insight: Confirms the relationship between perceived energy and actual audio intensity.

```python
sns.regplot(data= sample_data.sample(frac =0.05), y= "loudness", x= "energy", scatter_kws={"color": "blue"}, line_kws={"color": "red"}, ci=None).set(title= "Regression Plot- Loudness Vs Energy Correlation")
```

**_Interpretation of the Output_**

* A cloud of points representing songs.
* A red regression line showing the average trend.

**_Typical pattern:_**
* The line slopes upward, showing a positive correlation.
* Meaning: as energy increases, loudness also tends to increase.

**_Statistical intuition:_**
* This supports that energetic tracks (like EDM, rock, or pop) are produced with louder mixing.
* Quieter or lower-energy tracks (like acoustic or classical music) have lower loudness values.

### Number of Songs Released Each Year



```python
Years = sp_tracks["release_date"].dt.year # Extracts the year from release_date into a series Years, and prints it.
Years 
```

```python
sns.displot(Years, discrete= True, color= "b").set(title="No of Songs Per Year")
```

```python
sp_tracks["Years"] = sp_tracks["release_date"].dt.year
year_track_count= sp_tracks[["id", "Years"]].groupby("Years").agg("count") #Creates a new Years column and then counts how many id entries exist for each year.
```

```python
year_track_count.reset_index().sort_values("id", ascending= False)[:5] # Aggregates songs per year (sorted chronologically) and plots them.
```

```python
from plotly.offline import iplot, init_notebook_mode
init_notebook_mode(connected=True)
import plotly.express as px
```

```python
songs_per_years = sp_tracks["Years"].value_counts().sort_index()
iplot(px.bar(songs_per_years, 
       title='Number of songs per year', 
       text_auto=True,
       labels=dict(index='Years',value='Songs'),
       color_discrete_sequence=['#801080']).update_xaxes(type='category'))
```

**_Explanation_**

* sp_tracks["Years"] is derived from the release_date column.
* value_counts().sort_index() counts how many tracks were released each year.
* The code uses Plotly Express (px.bar) to make an interactive bar chart.
* text_auto=True adds labels on each bar.
* Purple color (#801080) adds a consistent aesthetic.

**_Interpretation_**

* x-axis: Release years
* y-axis: Number of songs released
* Expect a rising trend in later years, since Spotify catalogs more modern music.

*Insight: The dataset is richer in modern-era songs — so analyses will be biased toward recent releases unless controlled*
